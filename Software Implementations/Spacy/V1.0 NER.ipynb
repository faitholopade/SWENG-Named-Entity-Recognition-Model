```python
# import necessary libraries
import spacy
import re

# load a pre-trained English language model from spacy library
nlp = spacy.load("en_core_web_sm")

# define sample text as a string
text = "I talked to John Smith from XYZ Corp yesterday. His email is john.smith@example.com. " \
       "If you have any questions, please email support@example.com."

# process the text using the pre-trained model and generate a Doc object
doc = nlp(text)

# initialize variables to store the customer name and email
customer_name = None
email = None

# iterate over each named entity in the Doc object
for ent in doc.ents:
    # if the entity is a person, store the person's name in the customer_name variable
    if ent.label_ == "PERSON":
        customer_name = ent.text
    # if the entity is an email address, store the email address in the email variable
    elif ent.label_ == "EMAIL":
        email = ent.text

# if the customer_name variable is not None, print the customer's name
if customer_name is not None:
    print("Customer name:", customer_name)

# if the email variable is not None, print the email address
if email is not None:
    print("Email:", email)
# if the email variable is still None, use a regular expression pattern to search for email addresses in the text
else:
    # define a regular expression pattern to match email addresses
    email_pattern = r"[a-zA-Z0-9._+-]+@[a-zA-Z0-9._+-]+\.[a-zA-Z0-9._+-]+"
    # search for all matches of the pattern in the text
    matches = re.findall(email_pattern, text)
    # if matches are found, assign the first email address to the email variable and print it
    if matches:
        email = matches[0]
        print("Email:", email)
```

```python
import spacy
import re

# Load the spaCy model for English
nlp = spacy.load("en_core_web_sm")

# Define the input text that we want to analyze
text = "I talked to John Smith from XYZ Corp yesterday. His email is john.smith@example.com. " \
       "If you have any questions, please call 555-123-4567 or email support@example.com."

# Process the text using the spaCy model to extract named entities
doc = nlp(text)

# Initialize variables to store the extracted customer name, email address, and phone number
customer_name = None
email = None
phone = None

# Iterate over the named entities extracted from the text
for ent in doc.ents:
    # If the entity is labeled as an email address, store it in the email variable
    if ent.label_ == "EMAIL":
        email = ent.text
    # If the entity is labeled as a phone number, store it in the phone variable
    elif ent.label_ == "PHONE":
        phone = ent.text

# If an email address has been found, print it out and exit
if email is not None:
    print("Email:", email)

# If no email address has been found, try to extract one using a regular expression
else:
    # Define a regular expression to match email addresses
    email_pattern = r"[a-zA-Z0-9._+-]+@[a-zA-Z0-9._+-]+\.[a-zA-Z0-9._+-]+"
    # Use the re.findall method to find all matches of the email pattern in the text
    matches = re.findall(email_pattern, text)
    # If at least one match is found, assign the first match to the email variable and print it out
    if matches:
        email = matches[0]
        print("Email:", email)

    # Define a regular expression to match customer names
    name_pattern = r"[A-Z][a-z]*\s[A-Z][a-z]*"
    # Use the re.findall method to find all matches of the name pattern in the text
    matches = re.findall(name_pattern, text)
    # If at least one match is found, assign the first match to the customer_name variable and print it out
    if matches:
        customer_name = matches[0]
        print("Customer name:", customer_name)

# If a phone number has been found, print it out and exit
if phone is not None:
    print("Phone:", phone)

# If no phone number has been found, try to extract one using a regular expression
else:
    # Define a regular expression to match phone numbers
    phone_pattern = r"\b\d{3}[-.]?\d{3}[-.]?\d{4}\b"
    # Use the re.findall method to find all matches of the phone pattern in the text
    matches = re.findall(phone_pattern, text)
    # If at least one match is found, assign the first match to the phone variable and print it out
    if matches:
        phone = matches[0]
        print("Phone:", phone)
```
